Import Libs

```
import seaborn as sns
import pandas as pd
```

Crate dataframe from CSV file provided

Show the shape and column information of the data

```
path = '/content/drive/My Drive/CSC5542/Lab1/Data.csv'
df = pd.read_csv(path)

df.shape

df.dtypes
```

Print out column names of data

```
cols = df.columns

for i in cols:
  print(i)
```

Distribution of diagnosises, M = Malignant, B= Benign


```
sns.countplot(x = 'diagnosis', data = df)
```

Create sub dataframes of data for Malignant Cancers and for Benign Cancers

```
Malignant = df.loc[df['diagnosis'] == 'M']
Benign = df.loc[df['diagnosis'] == 'B']
```

Relationship between the perimeter of the cancer and the symmetry of the cancer


```
sns.relplot(x='perimeter_mean', y='symmetry_mean', hue='diagnosis', data= df)
```


Relationship between the area of the cancer and the symmetry of the cancer

```


sns.relplot(x='area_mean', y='symmetry_mean', hue='diagnosis', data= df)
```

Symmetry distribution of Benign and Malignant Cancers

```

sns.distplot(Malignant.symmetry_mean)
sns.distplot(Benign.symmetry_mean)
```

Relationship between the perimeter and area of the cancer

```
sns.relplot(x='perimeter_mean', y='area_mean', hue='diagnosis', data= df)
```

Relationship between perimeter and the number of concave points

```
sns.relplot(x='perimeter_mean', y='concave points_mean', hue='diagnosis', data= df)
```

Distribution of the area for Malignant and Benign cancers

```


sns.distplot(Malignant.area_mean)
sns.distplot(Benign.area_mean)

```

Distribution of the symmetry for Malignant and Benign cancers

```
sns.distplot(Malignant.symmetry_mean)
sns.distplot(Benign.symmetry_mean)
```

Distribution of the concave points for Malignant and Benign cancers

```
sns.distplot(Malignant['concave points_mean'])
sns.distplot(Benign['concave points_mean'])
```
